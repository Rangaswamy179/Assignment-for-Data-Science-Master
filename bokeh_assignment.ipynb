```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
#we can customize the appearance of a Bokeh plot by using the axis_label,title,legend_label
import numpy as np
x=[-6,-5,-4,-3,-2,-1,0,2,3,4,5,6,7,8,9,11]
y=np.sin(x)
output_file('test.html')
p=figure(title='Sine wave')
p.xaxis.axis_label='x axis sine wave'
p.yaxis.axis_label='y axis sine wave'
p.line(x,y,color='red',legend_label='sine wave data')
show(p)
```

```python
Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
```

```python
To embed a Bokeh plot into a web page or dashboard using Flask or Django, you can follow these steps:

Create a Bokeh plot using the Bokeh library in Python.
Use the json.dumps method to convert the Bokeh plot object into a JSON string.
In your Flask or Django application, create a route that will serve the web page or dashboard where you want to embed the Bokeh plot.
In the route, pass the Bokeh plot JSON string as a parameter to the Jinja template that will render the web page or dashboard.
In the Jinja template, create a div element with a unique ID that will be used to render the Bokeh plot.
Use JavaScript to load the Bokeh library and create a Bokeh plot using the JSON string and the unique ID of the div element.
Here's some sample code for embedding a Bokeh plot in a Flask application:
```

```python
from flask import Flask, render_template
from bokeh.plotting import figure
import json

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    p = figure()
    p.line([1, 2, 3], [4, 5, 6])
    
    # Convert the Bokeh plot to a JSON string
    plot_json = json.dumps(p.to_json())
    
    # Render the web page and pass the Bokeh plot JSON string to the Jinja template
    return render_template('index.html', plot_json=plot_json)

if __name__ == '__main__':
    app.run()
```
