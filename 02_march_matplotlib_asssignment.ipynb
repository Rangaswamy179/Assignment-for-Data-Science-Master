```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('x axis data')
plt.ylabel('y axis data')
plt.title('Scatter Plot Graph')
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A bar plot is one type of graph that represents numerical and categorical data with rectangular bars with respective height and
length with propotional to the values
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(5,3))
plt.bar(company,profit,color ='#753B8A')
plt.xlabel('represent company names')
plt.ylabel('represent company profit')
plt.title('bar graph represents company and thier profit')
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(company,profit,color ='#3B8A43')
plt.xlabel('represent company  profit ')
plt.ylabel('represent company names')
plt.title(' horizental bar graph represents company and thier profit')
plt.grid()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
Box plot is also Known as whisker plot is created to display summary of set values that contains
minimum,first quartile, median, third quartile and maximum.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box=[box1,box2]
plt.figure(figsize=(5,3))
plt.boxplot(box,patch_artist=True)
plt.show()
```
